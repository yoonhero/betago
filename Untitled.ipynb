```python
for file in files:
    try:
        with open(file, "r") as f:
            text = f.read().splitlines()[0]

        winner = text.split(",")[-1].strip()
        if winner != "0":
            print(text)
    except: pass
```

```python
import os
board = GoMuKuBoard(nrow=20, ncol=20, n_to_win=5)
for pos in text[1:-4]:
    x, y = pos.split(",")[:-1]
    x, y = int(x), int(y)
    board.set(x, y)
    os.system('clear')
    print(board)
```
